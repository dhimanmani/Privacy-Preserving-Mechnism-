```python
import traci
import sumolib

# Path to your SUMO configuration file
sumo_cfg = "/Users/manidhiman/sumo/map1.sumocfg"

# setting up traci for sumo
def start_sumo():
    sumo_binary = sumolib.checkBinary("sumo-gui")  # replace "sumo" if GUI is not needed
    traci.start([sumo_binary, "-c", sumo_cfg]) # start SUMO and establishes a traci connection
    print("SUMO started with TraCI")

# run a basic simulation loop
def run_simulation():
    try:
        start_sumo()

        step = 0
        while step < 1000:  # number of simulation steps
            traci.simulationStep()  # advance one step in SUMO
            vehicle_ids = traci.vehicle.getIDList()  # list of vehicle IDs
            
            for vehicle_id in vehicle_ids:
                speed = traci.vehicle.getSpeed(vehicle_id)  # vehicle speed
                road_id = traci.vehicle.getRoadID(vehicle_id)  # current road
                print(f"Vehicle {vehicle_id} is on road {road_id} moving at {speed} m/s")

            step += 1
    except Exception as e:
        print(f"An error occurred: {e}")
    finally:
        traci.close()  # close the TraCI connection
        print("Simulation ended")

# Run the simulation
run_simulation()

```
